```python
import pandas as pd
import numpy as np
print("Hello Capstone Project Course!")

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

### Importing Data from CSV

```python
df = pd.read_csv('data/Mumbai_Rent.csv')
df.head()
```

```python
df.describe(include='all')
```

### Reducing the DataFrame to features used in the study

```python
list_of_features=['area','bathroom_num','bedroom_num','floor_count','floor_num','furnishing','locality','user_type','price']
df_reduced = df[list_of_features]
df_reduced.head()
```

## Data Cleaning
---
### Any rows with empty entries will be dropped

```python
df_reduced.dropna(axis=0, how='any', inplace=True)
len(df_reduced)
```

### The datatype of all features will be fixed
Eg: Area is float 
    Bathroom_num, bedroom_num, floor_count, floor_num is int

```python
df_reduced.dtypes
```

```python
df_reduced[['area','price']]=df_reduced[['area','price']].astype('float')
df_reduced[['bathroom_num','bedroom_num','floor_count','floor_num']] = df_reduced[['bathroom_num','bedroom_num','floor_count','floor_num']].astype('int')
df_reduced.dtypes
```

### Creating dummy variables for categorical values

```python
df_locality = df_reduced.groupby('locality').filter(lambda x : len(x)>500)
df_dummy = pd.get_dummies(df_locality[['furnishing','locality','user_type']])
df_dummy.rename(columns={
    'furnishing_Furnished':'Furnished', 
    'furnishing_Semi-Furnished':'Semi-Furnished',
    'furnishing_Unfurnished':'Unfurnished',
    'user_type_Agent':'Agent',
    'user_type_Builder':'Builder',
    'user_type_Owner':'Owner'}, inplace=True)
df_dummy.head()
```

### Normalizing Data to reduce model bias

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
columns = ['area','bathroom_num','bedroom_num','floor_num','floor_count']
```

```python
temp_df = df_locality[['area','bathroom_num','bedroom_num','floor_count','floor_num']]
x = temp_df.values #returns a numpy array
x_scaled = min_max_scaler.fit_transform(x)
temp_df = pd.DataFrame(x_scaled)
temp_df.columns = ['area','bathroom_num','bedroom_num','floor_num','floor_count']
df_dummy.reset_index(inplace=True)
df_dummy.drop(['index'], axis=1, inplace=True)
df_clean = pd.concat([temp_df, df_dummy], axis=1)
# df_clean.drop(['index'], axis=1)
df_clean.head()
# print(len(temp_df),len(df_dummy))
# temp_df.head()
```

### Finding Pearson Correlation between features and target

```python
df_complete = df_clean
df_locality.reset_index(inplace=True)
df_complete['price'] = df_locality['price']
from scipy import stats
cols = ['area','bathroom_num','bedroom_num','floor_num','floor_count']
coeff = []
p_val = []
for col in cols:
    pearson_coef, p_value = stats.pearsonr(df_complete[col], df_complete['price'])
    coeff.append(pearson_coef)
    p_val.append(p_value)
    print('For feature {}:\nPearson Coefficient is {}\nP-Value is {}\n---------\n'.format(col,pearson_coef,p_value))
```

### Visualization of Pearson Correlation

```python
df_graph_pearson = pd.DataFrame({
    "Feature" : cols,
    "Pearson-Coefficient" : coeff,
    "P-Value" : p_val
})
plt.figure(figsize=(18,5))
g = sns.barplot(x="Pearson-Coefficient", y="Feature", data=df_graph_pearson, label="Pearson-Coefficient", color='c')
for index, row in df_graph_pearson.iterrows():
    g.text(row['Pearson-Coefficient'],row.name, round(row['Pearson-Coefficient'],6), color='black')
plt.savefig('graphs/Pearson-Coefficient.pdf')
plt.show()
```

### Building the Linear Regression model

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
regr = linear_model.LinearRegression()
x = df_clean
# x.drop(['price'],axis=1,inplace=True)
y = df_locality['price']
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.4)
regr.fit (x_train, y_train)
```

```python
from sklearn import metrics
yhat = regr.predict(x_test)
```

```python
score = regr.score(x_test,y_test)
score
```

### Plotting distribution graph for Model Evaluation

```python
ax1 = sns.distplot(y_test, hist=False, color="r", label="Actual Value")
sns.distplot(yhat, hist=False, color="b", label="Fitted Values" , ax=ax1)


plt.title('Actual vs Fitted Values for Price')
plt.xlabel('Price')
plt.ylabel('Distribution')
plt.savefig('graphs/DistGraph.pdf')
plt.show()
```

### Model Coefficients and Intercept

```python
coeff_name = list(df_clean.columns)
coeff_name.insert(0, 'Intercept')
coeff_val = list(regr.coef_)
coeff_intercept = regr.intercept_
coeff_val.insert(0, coeff_intercept)
df_coeff = pd.DataFrame({
    'Parameters' : coeff_name,
    'Value' : coeff_val
})
df_coeff.set_index('Parameters', inplace=True)
df_coeff
```
